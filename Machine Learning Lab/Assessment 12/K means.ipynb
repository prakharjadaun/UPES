```python
from sklearn.datasets import make_classification
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt
```

```python
x,t = make_classification(100,5,random_state=25)
```

```python
print(x.shape)
```

```python
# before we implement the k-means and assign the centers of our data, we can also make a quick analyze to find the optimal number (centers) of clusters using Elbow Method.
#  Collecting the distortions into list
distortions = []
K = range(1,10)
for k in K:
    kmeanModel = KMeans(n_clusters=k)
    kmeanModel.fit(x)
    distortions.append(kmeanModel.inertia_)
# Plotting the distortions
plt.figure(figsize=(16,8))
plt.plot(K, distortions, 'bx-')
plt.xlabel('k')
plt.ylabel('Distortion')
plt.title('The Elbow Method showing the optimal clusters')
plt.show()
```

```python
# we can observe that the “elbow” is the number 5 which is the optimal clusters (center) in this case

model = KMeans(n_clusters=5, random_state=80)
# Fit into our dataset fit
kmeans_predict = model.fit_predict(x)
```

```python
print(kmeans_predict)
```

```python
#visualizing the clusters
plt.scatter(x[kmeans_predict == 0, 0], x[kmeans_predict == 0, 1], s = 100, c = 'red')
plt.scatter(x[kmeans_predict == 1, 0], x[kmeans_predict == 1, 1], s = 100, c = 'blue')
plt.scatter(x[kmeans_predict == 2, 0], x[kmeans_predict == 2, 1], s = 100, c = 'green')
plt.scatter(x[kmeans_predict == 3, 0], x[kmeans_predict == 3, 1], s = 100, c = 'orange')
plt.scatter(x[kmeans_predict == 4, 0], x[kmeans_predict == 4, 1], s = 100, c = 'pink')
```
